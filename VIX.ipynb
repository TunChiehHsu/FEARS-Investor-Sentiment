#### Michigan Index of Consumer Sentiment

```python
mich_data = pd.read_csv('/Users/mueric35/Desktop/Sentiment-and-Marktet-Analysis/mich.csv')
```

```python
mich_data['date'] = mich_data['Month'].astype(str) + '-' + mich_data['YYYY'].astype(str)
mich_data['date'] = pd.to_datetime(mich_data["date"],format = '%B-%Y')
start = datetime.strptime('2010-01-01', "%Y-%m-%d")
```

```python
mich_data= mich_data[mich_data['date'] >= start]
count =  np.array(mich_data['ICS_ALL'])
count.astype(int)
pp = Series(count, index= mich_data.date)
plt.figure(figsize = (20,8))
plt.title('Michigan Index of Consumer Sentiment')
pp.plot(color = 'darkblue')
print('Michigan Index of Consumer Sentiment')
plt.show()
```
